# Introduction to pandas
- Pandas is a Python module designed to handle tabular data.
- Tabular data - data that has a shape like a table, with rows and columns.
- By convention, we use the short alias `pd` for pandas

```python
# importing the package(s) we want to use
import pandas as pd
```

## Create data
There are two core objects in pandas: the [**DataFrame**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) and [**Series**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html). In this exercise, we will create some sample data and also read from an existing file.

### **Series**  
A series is similar to a list and can be created by nothing more than a list. It has 2 additional attributes other than the list content; index and datatype


It is different than a list and has two additional attributes:
1.   An explicit index
2.   Type dtype (data type).

A dtype can be integer, float, and object. int64 - in Python which stands for 64 bit integer. The 64 refers to the memory allocated to store data in each cell which effectively relates to how many digits it can store in each “cell”

```python
# Converting a list into a pandas series
mylist = ['a','b','c','d']
my_series = pd.Series(mylist)
# print(my_series.dtype)
my_series
```

We can also create a pandas Series with a *list of strings*

```python
sList = ['ABC','DEF','GHI']

# Creating a series
s = pd.Series(sList)
print(s.dtype)
```

Noticed that the dtype is `object`, not `str` as expected.   
This is because panadas uses object by default for strings. You can explicitly set the dtype to string by using the dtype argument.

```python
# using the dtype attribute to convert the obj to string
pd.Series(sList, dtype="string")
```

A Series is, in essence, a single column of a DataFrame. So you can assign column values to the Series, using an index parameter via `index`. However, a Series does not have a column name, it only has one overall name, assigned as `name`.

```python
revenue = [30, 38, 58]
years = ['2018 Sales', '2019 Sales', '2020 Sales']

# using the name and index attribute
pd.Series(revenue, index=years, name="Product A")
```

### **DataFrame**   
We can think of a Series as a list and a DataFrame as a table,. By convention, we use a generic name `df` for DataFrame objects. A DataFrame is made of multiple Series combined together.
   
We are using the `pd.DataFrame()` to generate these DataFrame objects. To convert a dictionary into a DataFrame object, pandas will use the **keys** into the column names and the **value** will be the data entry for the columns.

```python
my_dict = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

# Convert a dictionary into data frame
my_df = pd.DataFrame(mydataset)

print(my_df)
```

Additionally, you can use multiple lists to construct a DataFrame object. The syntax for declaring a new one is a dictionary whose **keys** are the **column names** (Names and Scores in this example), and whose **values** are a **list of entries**. This is the standard way of constructing a new DataFrame, and the one you are most likely to encounter.

```python
# create a df with list and providing a custom index by using a dictionary adjacent structure

list1 = ['a','b','c', 'd']
list2 = [1,3,4,7]

df = pd.DataFrame({"Names": list1, "Scores": list2})
df
```

The dictionary-list constructor assigns values to the column labels, but just uses an ascending count from 0 (0, 1, 2, 3, ...) for the **row labels**.

A list of row labels known as an **Index** can be used in a DataFrame. We can assign values to it by using an index parameter in our constructor.

```python
# Specify a custom index using the index parameter

productList = ['Product A', 'Product B', 'Product C', 'Product D']
pd.DataFrame({"Names": list1, "Scores": list2}, index=productList)
```

### Index-based selection
The Series and the DataFrame are intimately related. It's helpful to think of a DataFrame as actually being just a bunch of Series "glued together". We can use indexing operators, `iloc` and `loc`, to select for data.
  
Pandas has its own indexing operators, `iloc` and `loc`.
- `.iloc[ ]`: index-based selection that selects data based on its numerical position in the data
- `.loc[ ]`: primarily label based

&nbsp;  
**Indexing in pandas**  
- [start:stop:skip] #stop is exclusive [1:3] --> index 1 & 2  
- [1::2, :] #from the 2nd row all the way to the end every other row, and all the columns  
- [:, :3] #all rows, from the beginning to the 3rd column (0, 1, 2)

### Working with missing values
Different methods to work with missing values include dropna, isna, and fillna

#### **dropna():**

Purpose: Removes missing values from a DataFrame or Series.

Usage: df.dropna(axis=0, how='any')  


#### **isna():**

Purpose: Identifies missing values, returning a boolean mask indicating their location.

Usage: df.isna()

#### **fillna():**

Purpose: Fills missing values in a DataFrame or Series with a specified value or method.

Usage: df.fillna(value=0, method=None)
